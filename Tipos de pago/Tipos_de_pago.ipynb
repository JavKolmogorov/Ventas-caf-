#Importación de las librerías


```python
import pandas as pd
import matplotlib.pyplot as plt
```

#Importación del dataset


```python
cafe_df = pd.read_csv('index.csv')
cafe_df.head()
```

#Porcentaje de tipos de pago


```python
porcentaje_df = cafe_df.groupby("cash_type")["cash_type"].count()/len(cafe_df)*100
plt.pie(porcentaje_df, labels=porcentaje_df.index, autopct='%1.1f%%')
plt.show()
```
